```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#Load and read the data
df=pd.read_csv('/content/Weather Dataset - Trial Activity DataSet.csv.csv')
df.head()
```

```python
#Dataset information
df.info()
```

```python
#Check for null values
df.isnull().sum()
```

```python
#Drop the null values
df.drop('is_holiday',axis=1,inplace=True)
```

```python
#Drop unnecessary columns
df.drop(['rain_p_h', 'snow_p_h'], axis=1, inplace=True)
```

```python
df.columns
```

```python
sns.pairplot(df)
```

```python
#Bargraph
sns.barplot(x="weather_type",y="air_pollution_index",data=df,palette="hls")
plt.title("Weather type vs Air Pollution Index")
plt.xlabel("Weather type")
plt.ylabel("Air Pollution Index")
plt.show()
```

```python
#Countplot
sns.countplot(x="visibility_in_miles",data=df, palette='deep')
plt.title("Count of Visibility in Miles")
plt.xlabel("Visibility in Miles")
plt.ylabel("Count")
plt.show
```

```python
#histogram/distplot

sns.displot(x='temperature',y='dew_point',data=df,palette='Blues')
plt.title("Temperature vs DewPoint")
plt.xlabel("Temperature")
plt.ylabel("Dew Point")
plt.show()
```

```python
#Scatter Plot
sns.scatterplot(x='humidity',y='temperature',data=df,palette='magma')
```

```python
sns.swarmplot(x="temperature",y="dew_point", data=df)
```

```python
sns.stripplot(x='humidity',y='temperature',data=df,palette='magma')
```

```python
#displot
sns.displot(df['temperature'],kde=True)
```
